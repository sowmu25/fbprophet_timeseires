```python
import yfinance as yf
import pandas as pd
```

```python
tsla=yf.Ticker("TSLA")
hist=tsla.history(period='3y')
hist.to_csv("tesla.csv")
```

```python
tesla=pd.read_csv("tesla.csv")
tesla
```

```python
from sklearn.ensemble import IsolationForest
```

```python
tesla.info()
```

```python
model=IsolationForest()
model.fit(tesla.iloc[:,1:])
tesla["anomaly"]=model.predict(tesla.iloc[:,1:])
tesla["snomaly_score"]=model.decision_function(tesla.iloc[:,1:])
```
